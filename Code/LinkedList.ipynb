# Linked List

* **Extension of list, but definitely not an array** 

* It is an another common data structure that complements the array data structure 

* A linked list is **made up of nodes that hold only two pieces of information: the data they have to store, and the memory address of the next node in the list.**

* The linked list **begins with the head node and ends with the tail node.** The tail node holds a null value in the place of the next nodeâ€™s memory address.

* A linked list is a dynamic data structure. The number of nodes in a list is not fixed and can grow and shrink on demand. Any application which has to deal with an unknown number of objects will need to use a linked list.


## Operations in Linked List 

* Insertion : 

* Removal/Deletion 

#### Insertion and Deletion in Linked List takes constant amount of time i.e. O(1)

### Similarity of Linked List and Array 

* Both are linear data structure and stores element in a linear fashion 

### Dis-similarities of Linked List and Array 

* Linked list is dynamic data structure (easily can be expanded) . While array is static data structure(always fixed and size cannot be changed) 

* Arrays stored elements in contiguous locations. While linked list doesn't store them in contiguous locations; instead, they are scattered everywhere in memory, which is connected to each other using nodes

* Both **linked list and array stores different kind of information**: 

* * Each element in array has: 
            i. Value 
            ii. Index 
            
* * Each element in linked list has:
            i. Value
            ii. Address/reference of next element 
            
### Problems in Array

* Adding and removing the elements in an array can be really complicated. While, adding and removing elements in linked list is comparatively effortless as compare to arrays 

* Array is static data structure therefore size of array needs to be defined before, and it is always fixed. This can cause problems during the implementation of stack and queues

* Linked list is dynamic data structure, its size can be modified during runtime

### Problems in Linked List 

* One disadvantage of a linked list against an array is that it does not allow direct access to the individual elements. If you want to access a particular item then you have to start at the head and follow the references until you get to that item.  **(Access time of element : Linear time O(n))**

* Another disadvantage is that a linked list uses more memory compare with an array - we extra 4 bytes (on 32-bit CPU) to store a reference to the next node.

### Advantages of Linked List

* Insertion and deletion takes constant time i.e. O(1) 

### Applications:

* Choice for underlying data structure such as **Stacks and Queues**
* Anytime we need **fast insertions and deletion, but random access less important** 


### Implementation :

* Since Linked List needs the concept of pointers. In python, we do not have pointers. Therefore we use similar concept called **Nodes**

### Note: 

**Linked List are mostly used to implement Stacks and Queues**

### Types of Linked List

i. Single Linked List: 

    * Previous element stores the reference address of the next element
    * Last element in the list stores the reference address as "null"

ii. Double Linked List:

    * A doubly linked list is a list that has two references, one to the next node and another to previous node.

iii. Circular Linked List:  

    * last node of the list points back to the first node (or the head) of the list.

# Nodes

Since current element needs to point to the address of the next element. Therefore, we need to use 
the concept of **pointers**. In python, we do not have **pointers**, but we can use concept of **Nodes**

source: https://www.tutorialspoint.com/python/python_nodes.htm


## Possible Operations: 

1. Creation 
2. Traversing
3. Insertion 
4. Removing

### 1. Creation of node Elements (same step for linked list also)

```python
# Creation of node elements
class Node:
    
    def __init__(self, day=None):
        
        # each node has two properties: 1. Node name 2. Next element address (name)
        self.day = day
        self.next_day = None

class LinkedList: 
    
    def __init__(self, head_element=None):
        
        self.head_e = head_element
    
    
    def traverse_element(self, specific_node=None):
        
        tmp_e = self.head_e
        while tmp_e:
            
            print (tmp_e.day)
            # if user want to transverse till some element, we break loop there
            if (tmp_e.day == specific_node):
                break
            tmp_e = tmp_e.next_day
            
    '''
    To remove element, user needs to specify name of node 
    '''
    
    def remove_element(self, node_name=None):
        
        if node_name == None:
            print ("no name of element is specified")
            return 
        
        tmp = self.head_e

        # if remove element is head element
        if (tmp is not None):
            if tmp.day == node_name:
                self.head_e = tmp.next_day
                tmp = None
                return 
            
        
        while (tmp.next_day is not None):
            
#             print ("node ", tmp.day)
            if (tmp.day == node_name):
                break
            
            prev = tmp
            tmp = tmp.next_day
        
        print ("prev ", prev.day)
        print ("tmp ", tmp.day)
        # if key not present in linked list
        if tmp == None:
            print ("key not present in list")
            return 
        
        # change the link
        prev.next_day = tmp.next_day
 
  
    # add element not in end/beginning
    '''
    This involves changing pointer of a specific node to point to a new node
    Params: pass new node and name of existing node
    '''
    
    def add_in_between(self, existing_node_name=None, new_node_name=None):
        
        new_element = Node(new_node_name)
        
        tmp_e = self.head_e
        
        if (existing_node_name == None or new_node_name == None):
            print ("please provide sufficient arguments ")
            return
        
        while tmp_e:

            if tmp_e.day == existing_node_name:

                tmp_pointer = tmp_e.next_day # this stores existing node pointing to next node
                tmp_e.next_day = new_element
                new_element.next_day = tmp_pointer
                return

            tmp_e = tmp_e.next_day

    
    def add_element_ending(self, new_node_name=None):
        
        new_element = Node(new_node_name)
        
        if new_node_name == None:
            print ("not sufficient arguments provided")
            return 
        
        if self.head_e == None:
            self.head_e = new_element
            return
        
        tmp_e = self.head_e
        while tmp_e:
            
            if (tmp_e.next_day is None):
                tmp_e.next_day = new_element
                return
           
            tmp_e = tmp_e.next_day
            
            
    def add_element_beginning(self, new_node_name=None):
        
        new_element = Node(new_node_name)
        new_element.next_day = self.head_e
        self.head_e = new_element
    
d1 = Node("Monday")
d2 = Node("Tuesday")
d3 = Node("Thursday")
d1.next_day = d2
d2.next_day = d3

ll = LinkedList(d1) # head element assigned 
print ("Traverse element ")
ll.traverse_element()
print "--------------------------------"
print ("Add element in beginning")
ll.add_element_beginning("Sunday")
print ("traverse elements again after adding")
ll.traverse_element()
print "--------------------------------"
print "\n"
print ("add element in ending")
ll.add_element_ending("Friday")
print "--------------------------------"
print ("traverse elements again after adding element at ending")
ll.traverse_element()
print "--------------------------------"
print "\n"
print '------ Add new node in between ------'
ll.add_in_between('Tuesday', 'Wednesday')
print '----------------------------------'
print 'Traverse to check new node added in between'
ll.traverse_element()
```
